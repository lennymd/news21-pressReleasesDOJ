## Getting the comments we want

So we have 165 files and the paths are all stored in a list. Now we're going to access each file and extract the first two comments. We want the second comment, but we'll grab the first one just in case something goes odd. I'll put the comments in their own list.

```python
import time
out_list = []

for file in fileList:
    i = fileList.index(file)
    print(i, file)
    test_file = file
    with open(test_file, 'rb') as input_data:
        response = requests.post(calais_url, data=input_data, headers=headers, timeout=80)
        content = response.text
    #print("got a response")
    soup=BS(content,'html.parser')
    comments=soup.find_all(string=lambda text:isinstance(text,Comment))
    for comment in comments[0:2]:
        out_list.append(comment)
    #print("got the comments")
    #print("@@")
    time.sleep(1)
```

```python
city_list = []
person_list = []

for row in rows:
    if row[0].strip() == 'City':
        print(rows.index(row),'City')
        for i in range(1,len(row)):
            city_list.append(str(row[i].strip()))
    elif row[0].strip() == 'Person':
        print(rows.index(row),'Person')
        for i in range(1,len(row)):
            person_list.append(str(row[i].strip()))
    else:
        print(rows.index(row),'Found nothing')
        pass
```

```python
cityDict = {}
with open('cities.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    rows = [row for row in reader]
```

```python
citiesToTest = []
for row in rows:
    citiesToTest.append(row[0])
```

```python
len(rows), len(citiesToTest)
```

```python
for CITY in citiesToTest:
    city = CITY.lower()
    print(city)
    if city in cityDict:
        cityDict[city] += 1
    else:
        cityDict[city] = 1
```

```python
cityDict
```

This worked well enough, but I want the people more than I want the cities. So let's run it with the people.

```python
personDict = {}
with open('persons.csv', 'r') as f:
    reader = csv.reader(f, delimiter=',')
    rows = [row for row in reader]
    
personCheck = []
for row in rows:
    personCheck.append(row[0])
    
for PERSON in personCheck:
    person = PERSON.lower()
    if person in personDict:
        personDict[person] += 1
    else:
        personDict[person] = 1
```

```python
personDict
```

Now to sort these dictionaries by size. I care about the person one so let's do that first for the meeting I have in a bit

```python
for key, value in sorted(personDict.items(), key=lambda kv: (-kv[1], kv[0])):
    print("%s: %s" % (key, value))
```

Going to write these results to  text file:

```python
with open('countedPersons.txt','w+') as f:
    for key, value in sorted(personDict.items(), key=lambda kv: (-kv[1], kv[0])):
        line = "%s: %s" % (key, value) + '\n'
        f.write(str(line))
```
