```python
with open('pressReleases_people4.csv','r') as f:
    reader = csv.reader(f, delimiter=',')
    rows = [row for row in reader]

# get header columns
for i in range(0,len(rows[0])):
    print(i,rows[0][i])
```

```python
defDict = {}
proDict = {}
judgeDict = {}
vicDict = {}

for row in rows:
    release = row[0]
    person = row[1].strip()
    role = row[2].strip()
    title = row[3].strip()
    case = str(row[6])
    
    if role == "defendant":
        if case in defDict:
            defDict[case] = defDict[case]+'; '+ person
        else:
            defDict[case] = person
    elif role == "prosecutor":
        if case in proDict:
            proDict[case] = proDict[case]+'; '+ person
        else:
            proDict[case] = person        
    elif role == "judge":
        if case in judgeDict:
            judgeDict[case] = judgeDict[case]+'; '+ person
        else:
            judgeDict[case] = person        
    else: #role == victim
        if case in vicDict:
            vicDict[case] = vicDict[case]+'; '+ person
        else:
            vicDict[case] = person
            
```

I want to combine all the temp tables into an untidy version that is easier for people to read:

```python
with open('tidyPressReleases.csv','r') as f:
    reader = csv.reader(f, delimiter=',')
    rows = [row for row in reader]
```

Let's start with the first case: `caseGroup=='1'`

```python
masterList = []

for i in range(1,73):
    caseGroup = i

    releaseList = []
    defendantList = []
    prosecutorList = []
    judgeList = []
    victimList = []

    for row in rows[1:]:
        if row[4] == str(caseGroup):
            # we're inside the rightcase
            release = row[0]
            person = row[1]
            role = row[2]
            title = row[3]


            if release not in releaseList: releaseList.append(release) 

            if role=="defendant":
                #print(row)
                if person in defendantList:
                    pass
                else:
                    defendantList.append(person)
            elif role =="prosecutor":
                #print(row)
                fulltitle = person+', '+title
                if fulltitle in prosecutorList:
                    pass
                else:
                    prosecutorList.append(fulltitle)
            elif role =="judge":
                #print(row)
                if person in judgeList:
                    pass
                else:
                    judgeList.append(person)
            else:
                if person in victimList:
                    pass
                else:
                    victimList.append(person)
        else:
            #there is no case in this part.
            pass
    output = [str(caseGroup),defendantList,prosecutorList,judgeList,victimList,releaseList]
    masterList.append(output)

print('done looping')
```
